###  Single thread CPU-Latency

```python
%matplotlib inline 
#notebook
import IPython
import pandas as pd
import matplotlib.pyplot as plt

thread_path= "/Users/developer/PycharmProjects/serverless-function-placement/tests/FIB-Multi-Thread/"



def data_collection_process(thread_count, resource):
    thread_path= "/Users/developer/PycharmProjects/serverless-function-placement/tests/FIB-Multi-Thread/"
    thread="Thread-"+str(thread_count)
    resource=resource
    dataCollect=[{"num":20, "label":""}, 
             {"num":20, "label":"-KF"}, 
             #{"num":20, "label":"-RAND"},
             {"num":30, "label":""}, 
             {"num":30, "label":"-KF"},
             #{"num":30, "label":"-RAND"},
             {"num":35, "label":""}, 
             {"num":35, "label":"-KF"},
             #{"num":35, "label":"-RAND"}
            ]
    for item in dataCollect:
        file_path= "{}{}/{}/FIB-{}-2-Node{}/main_core_requestlog.csv".format(thread_path,
                                                                         thread,resource, 
                                                                         item['num'], item['label'])
        print(file_path)
        try:
            d_frame=pd.read_csv(file_path, nrows=300)
            latency=d_frame.iloc[:,2]
            item['latency']=latency
            print(d_frame.describe())
        except:
            continue
            
    return dataCollect


dataCollect= data_collection_process(1,'CPU')
# dataMemCollect = data_collection_process(1,'MEM')
# print(dataCollect)


fig1, (ax1) = plt.subplots() #1, 2, figsize=(10, 3)

# cpu plot
ax1.set_title('Latency Plot CPU Single Thread')
labels = ["{}{}".format(item['num'],item['label']) for item in dataCollect]
ax1.set_ylabel("Seconds")
box1=ax1.boxplot([item['latency'] for item in dataCollect], labels=labels, patch_artist=False);


#box2=ax2.plot([item['url'] for item in dataCollect], labels=labels, patch_artist=False); 

#df=pd.read_csv(file_path)
#df.groupby(df.index)[].nunique().plot(kind='bar')
#plt.show()




# mem plot
# ax2.set_title('Latency Plot Memory Single Thread')
# labels = ["{}{}".format(item['num'],item['label']) for item in dataMemCollect]
# ax2.set_ylabel("Seconds")
# box2=ax2.boxplot([item['latency'] for item in dataMemCollect], labels=labels, patch_artist=False); 



```
